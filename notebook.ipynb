```python
import kagglehub
import pandas as pd
import shutil
import os
```

```python
print("Downloading dataset...")
cache_path = kagglehub.dataset_download(
    "alessandrasala79/ai-vs-human-generated-dataset")

# define where you want to copy
local_path = r"C:\sajana_works\assignment-adl"

# copy dataset to local_path
shutil.copytree(cache_path, local_path, dirs_exist_ok=True)

print(f"Dataset successfully copied to: {local_path}")
print("Files in local folder:", os.listdir(local_path))
```

```python
if os.path.exists(cache_path):
    print(f"Removing cache folder: {cache_path}")
    shutil.rmtree(cache_path)
    print("Cache removed.")
else:
    print("Cache path does not exist.")
```
